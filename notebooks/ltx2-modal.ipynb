# **FP8_E4M3 Issue Fix (Optional)**

```python
import os
import torch

# RUN THIS CELL if you get a "type fp8e4nv not supported" error on A100 GPUs.

# 1. Force Triton to use A100 (SM80) compatible math kernels
os.environ["TRITON_OVERRIDE_ARCH"] = "80"

# 2. Disable the specific high-precision FP8 mode that causes crashes on non-H100 GPUs
os.environ["ENABLE_FP8_E4M3"] = "0"

print(f"Environment patched for A100 compatibility. Current Torch: {torch.__version__}")
```
